```python
import pandas as pd
per = pd.read_csv('PER.csv')
starters = pd.read_csv('Starters.csv')
```

```python
starters.head()
```

```python
per.head()
```

```python
# MPG = Minutes played per game by a player
per['MPG'] = per['MP'] / per['G']
```

```python
per = per[['Player','PER','MPG']]
```

```python
per.head()
```

```python
final = pd.merge(per, starters, on='Player', how='outer')
```

```python
final.head()
```

```python
final = final[final.MPG >6.09]
final = final[['Player','PER','Team']]
final = final.dropna()
final.head()
```

```python
mean_per = final.mean()
mean_per= float(mean_per)
mean_per
```

```python
final.describe()
```

```python
margin = final.groupby("Team", as_index=False)["PER"].min()
margin['Gap'] = abs(margin['PER'] - mean_per)
margin
```

```python
maxmargin = final.groupby("Team", as_index=False)["PER"].max()
maxmargin['Gap'] = (margin['PER'] - mean_per)
maxmargin
```

```python
#Find PER divide between best and worst player
divide = final.groupby("Team", as_index=False)["PER"].max()
divide['Divide'] = divide['PER'] - margin['PER']
divide
```

```python
gap = margin[['Team','Gap']]
gap
```
